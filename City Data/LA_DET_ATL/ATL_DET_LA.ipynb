```python
import datetime

import pandas as pd
df_ATL = pd.read_csv("ATL_crime_data.csv")
df_DET = pd.read_csv("detroit_data.csv")
df_LA = pd.read_csv("LA_crime_data.csv")
```

```python
#convert date units to UNIX
```

```python
df_ATL.columns
```

```python
#narrowing down to necessary columns
df_ATL_edit = df_ATL[["Location 1", "CrimeDate", "Description", "CrimeCode"]]

df_ATL_edit.rename(columns={'Location 1': 'Location', 'CrimeDate': 'Date', 'Description': 'Indoor/Outdoor', 'CrimeCode': 'Crime Code'}, inplace=True)

df_ATL_edit.head()
```

```python
#changing date format to UNIX
df_ATL_edit["Unix Date"] = df_ATL_edit["Date"].map(lambda x:datetime.datetime.strptime(x, "%m/%d/%Y").timestamp())
df_ATL_edit['Unix Date'] = df_ATL_edit['Unix Date'].apply('astype',dtype='int64')

df_ATL_edit.head()
```

```python
df_ATL_edit.to_csv("ATL_Unix.csv")
```

```python
#same with Detroit
df_DET.columns
```

```python
#combining latitude + longitude columns into one "location" column
df_DET["Location"] = '(' + df_DET["LAT"].map(str) + ', ' + df_DET["LON"].map(str) + ')'

#same with offense
df_DET["Offense"] = df_DET["CATEGORY"] + ': ' + df_DET["OFFENSEDESCRIPTION"]
df_DET.head()
```

```python
#cleaning up Detroit df to only include necessary columns
df_DET_edit = df_DET[["Location", "Offense", "INCIDENTDATE"]]

df_DET_edit.rename(columns={'INCIDENTDATE': 'Date'}, inplace=True)

df_DET_edit.head()
```

```python
#changing date format to UNIX
df_DET_edit["Unix Date"] = df_DET_edit["Date"].map(lambda x:datetime.datetime.strptime(x, "%m/%d/%Y").timestamp())
df_DET_edit['Unix Date'] = df_DET_edit['Unix Date'].apply('astype',dtype='int64')

df_DET_edit.head()
```

```python
df_DET_edit.to_csv("DET_Unix.csv")
```

```python
#LA
df_LA.columns
```

```python
#getting LA df down to necessary columns
df_LA_edit = df_LA[["Location 1", "DATE OCC", "Crm Cd", "Crm Cd Desc"]]

df_LA_edit.rename(columns={'Location 1': 'Location', 'DATE OCC': 'Date', 'Crm Cd Desc': 'Offense', 'Crm Cd': 'Crime Code'}, inplace=True)

df_LA_edit.head()
```

```python
df_LA_edit.to_csv("LA_Unix.csv")
```

```python
frames = [df_ATL_edit,df_LA_edit,df_DET_edit]
df_unordered = pd.concat(frames, ignore_index = True, sort=True)

df_unordered.head()
```

```python
df = df_unordered[["Date", "Crime Code", "Offense", "Location"]]

df.head()
```
